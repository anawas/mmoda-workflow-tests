```python
import matplotlib.dates as mdates
import pandas as pd

# We also generate an image output
angle = np.linspace(0, 2*np.pi, 100)
y = np.sin(angle)

dates = pd.date_range('2024-01-01', periods=len(y), freq="D")
fig, ax = plt.subplots()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y/%m/%d"))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax.xaxis.set_label("Observation date")
plt.gcf().autofmt_xdate()
plt.title("Simple sine function")
ax.grid()
ax.plot(dates, y)
plt.savefig("test.png")
result_image = PictureProduct.from_file("test.png", name="sine_wave")
```
